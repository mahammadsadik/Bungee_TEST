```
import pandas as pd
url = 'https://raw.githubusercontent.com/Bungeetech/internship-test/master/input/main.csv'
df = pd.read_csv(url)
fdf = df.COUNTRY.str.contains('USA') #filtering the data which has the country USA
newdf = df[fdf] 
newdf
```

```
first_lowest_df = newdf.groupby('SKU').PRICE.nth(0).to_frame(name = 'PRICE').reset_index() #shows the first lowest price 
second_lowest_df = newdf.groupby('SKU').PRICE.nth(1).to_frame(name = 'PRICE').reset_index() #shows the second lowest price
```

```
first_lowest_df
```

```
second_lowest_df
```

```
df3 = pd.merge(first_lowest_df,second_lowest_df,on='SKU') #merging both the data frames on SKU value
df3
```

```
df3.columns #finding the name of the columns
```

```
df3.rename(columns={'PRICE_x':'first_minimum_price', 'PRICE_y':'secon_minimum_price'}, inplace=True) #renaming the column names
df3
```
